```python
### settingup the path and readign all the files present in folder

## this is for aggregated transaction data

# folder_path=r"C:\Users\nagan\OneDrive\Desktop\Guvi\Project\PhonePe\Data Processing\Files\GitHub_Raw_Data\data\aggregated\transaction\country\india\state"
folder_path=output_path+"/data/aggregated/transaction/country/india/state/"
agg_state_list=os.listdir(folder_path)
agg_state_list[:10]

# print(folder_path)

########## Creatign Dictionary to store only required columns from all the file using for loop


data_dict={'State':[],'Year':[],'Quarter':[],'Transaction_Type':[],'Transaction_Count':[],'Transaction_Amount':[]}

## Looping through all the folders and files for each state

for i in agg_state_list: #  here i is state folder
    year_path=folder_path+i+"/"
    agg_yrs=os.listdir(year_path) # this list ill have all the years for each state data
    # print(agg_yrs)

    for year in agg_yrs:
        file_path=year_path+year+"/"  ## path for all the files present in the state folder
        agg_files=os.listdir(file_path)

        # print(agg_files)
        for k in agg_files:
            json_file=file_path+k  ## json file path
            # print(file)
            data=open(json_file,'r')
            d=json.load(data)
            for z in d['data']['transactionData']:
                Name=z['name']
                count=z['paymentInstruments'][0]['count']
                amount=z['paymentInstruments'][0]['amount']
                data_dict['Transaction_Type'].append(Name) #appending type amount and count to defined dictionary
                data_dict['Transaction_Count'].append(count)
                data_dict['Transaction_Amount'].append(amount)
                data_dict['State'].append(i)
                data_dict['Year'].append(year)
                data_dict['Quarter'].append(int(k.strip('.json')))


## creating the dataframe
agg_df=pd.DataFrame(data_dict)

display(agg_df)

#writing it to csv
agg_df.to_csv(f"{processed_data_path+'/'}aggregated_transaction_data.csv",index=False)
```
